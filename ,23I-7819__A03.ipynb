```python
#'Comp Net (CS-C/D)'
file_path = "TimeTable, FSC, Fall-2023.xlsx"

# Create a list of sheet names to iterate through
sheet_names = ["Monday", "Tuesday", "Wednesday", "Thursday"]

# Ask the user for the subject they are interested in
subject_to_find = input("Enter the subject you want to find: ")

# Iterate through each sheet
for sheet_name in sheet_names:
    print(f"Sheet: {sheet_name}\n")
    
    # Read the current sheet into a DataFrame with header=None
    df = pd.read_excel(file_path, sheet_name=sheet_name, header=None)
    
    # Delete the first 4 rows from the DataFrame and reset index
    df = df.iloc[4:].reset_index(drop=True)
    
    # Iterate through the DataFrame and display the time, room, and subjects
    for index, row in df.iterrows():
        if index == 0:
            # Header row, which contains the time slots, print as is
            print("\t".join(str(cell) for cell in row))
        else:
            # Data rows, extract time and room index from the first column
            time = row[0]
            room = index  # Room index is the row index
            subjects = [cell for cell in row[1:] if subject_to_find in str(cell)]
            if subjects:
                print(f"Time: {time}, Room Index: {room}, Subjects: {', '.join(subjects)}")
    
    print("\n" + "-"*50 + "\n")  # Separate each sheet's output
```
